```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
```

```python
students_performance = students_performance.rename(columns=
                                                   {"parental level of education":"parental_level_of_education",
                                                           "test preparation course": "test_preparation_course",
                                                           "math score": "math_score",
                                                           "reading score": "reading_score",
                                                           "writing score": "writing_score"})
```

```python
students_performance.head()
```

```python
students_performance.groupby(["gender", "race/ethnicity"], as_index=False) \
.aggregate({'math_score': 'mean', 'reading_score': "mean", 'writing_score': 'mean'}) \
.rename(columns={"math_score": 'mean_math_score', "reading_score": "mean_reading_score", "writing_score": "mean_writing_score"})
```

```python
students_performance.sort_values(["gender", "math_score"], ascending=False) \
.groupby("gender").head(5)
```

```python
students_performance["total_score"] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

```python
students_performance.drop(["total_score"], axis=1)
```

```python
concentrations = pd.read_csv("http://stepik.org/media/attachments/course/4852/algae.csv")
```

```python
concentrations.head()
```

```python
concentrations.groupby("genus")[["sucrose", "alanin", "citrate", "glucose", "oleic_acid"]].mean()
```

```python
concentrations.groupby("genus").aggregate({"sucrose":"mean", "alanin":"mean", "citrate":"mean", "glucose":"mean",
                                            "oleic_acid":"mean"})
```

```python
np.round(concentrations.loc[concentrations.genus == "Fucus"].groupby("genus").aggregate({"alanin":"min"}),2)
```

```python
np.round(concentrations.loc[concentrations.genus == "Fucus"].groupby("genus").aggregate({"alanin":"mean"}),2)
```

```python
np.round(concentrations.loc[concentrations.genus == "Fucus"].groupby("genus").aggregate({"alanin":"max"}),2)
```
